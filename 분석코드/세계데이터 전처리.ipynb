```python
#60211670 송지은
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import stats
from matplotlib import font_manager, rc
from decimal import Decimal
import matplotlib.pyplot as plt

data1 = pd.read_csv("../자료/WHO-COVID-19-global-data.csv")
data1
```

```python
data_1 = data1.groupby("Date_reported")['New_cases'].mean().astype(int)
data_1
```

```python
data_2 = data1.groupby("Date_reported")['Cumulative_cases'].mean().astype(int)
data_2
```

```python
data_3 = data1.groupby("Date_reported")['New_deaths'].mean().astype(int)
data_3
```

```python
data_4 = data1.groupby("Date_reported")['Cumulative_deaths'].mean().astype(int)
data_4
```

```python
data = pd.concat([data_1,data_2,data_3,data_4],axis=1)
data
```
